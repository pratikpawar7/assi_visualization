```python
'''
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:'''
'''
subplot() function : 
    The subplot() function in Matplotlib is used to create multiple plots in the same figure. 
    It takes three arguments: the number of rows, the number of columns, and the index of the current plot.
    
'''
'''
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
'''


import matplotlib.pyplot as plt
import numpy as np


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, axis = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))


axis[0, 0].plot(x1, y1, color='blue')
axis[0, 0].set_title('Line 1')

axis[0, 1].plot(x2, y2, color='red')
axis[0, 1].set_title('Line 2')

axis[1, 0].plot(x3, y3, color='green')
axis[1, 0].set_title('Line 3')

axis[1, 1].plot(x4, y4, color='purple')
axis[1, 1].set_title('Line 4')


plt.tight_layout()

plt.show()
```

```python
'''
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
'''
'''
bar plot:
    A bar plot is a type of plot in which rectangular bars are used to represent the data. 
    The length of each bar is proportional to the value of the corresponding data point. 
    Bar plots are used to display and compare the values of different categories of data
    
    
uses : 
    Bar plots are often used in scientific research, business analysis, and data visualization. 
    They provide a simple and effective way to summarize and communicate complex data to a broad audience.
    '''

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
```
